```python
import matplotlib.pyplot as plt
def plot_fraud_by_column(df, column):
    table = (
        df.groupby(column)['FraudFound_P']
          .value_counts(normalize=True)
          .unstack(fill_value=0)
    )

    table.plot(kind='barh')
    plt.title(f'Fraud rate by {column}')
    plt.xlabel('Proportion')
    plt.ylabel(column)
    plt.legend(['Not Fraud', 'Fraud'])
    plt.tight_layout()
    plt.show()
```

```python
plot_fraud_by_column(df,'Year')
```

```python
plot_fraud_by_column(df,'Sex')
```

```python
plot_fraud_by_column(df,'Month')
```

## Exploring the '0' in Age column and handling it

```python
df_clean_v1 = df.copy()
df_clean_v1["Age"].describe()
```

```python
df_clean_v1.loc[df_clean_v1["Age"] == 0].value_counts().count()
```

```python
median_age = df_clean_v1['Age'][df_clean_v1['Age'] != 0].median()
df_clean_v1['Age'] = df_clean_v1['Age'].replace(0, median_age)
```

## Binary Encoding

```python
from sklearn.preprocessing import LabelEncoder
binary_columns = []

for col in df_clean_v1.columns:
    if col != "FraudFound_P" and df_clean_v1[col].nunique() == 2:
        binary_columns.append(col)

binary_columns
```

```python
le = LabelEncoder()

for col in binary_columns:
    df_clean_v1[col] = le.fit_transform(df_clean_v1[col].astype(str))

    print(col, dict(zip(le.classes_, le.transform(le.classes_))))
```
